```python
with open('/content/insurance.csv',newline='\r\n') as insurance_csv_file:
# Create variables
  insurance_data_dict = {}
  ages = []
  sexes = []
  BMIs = []
  children = []
  smokers = []
  regions = []
  charges = []
  insurance_data_reader = csv.DictReader(insurance_csv_file)

# # Create dictionary from CSV
#   i=0
#   for row in insurance_data_reader:
#     insurance_data_dict[i] = row
#     i+=1
#   print(insurance_data_dict)
# Create variable lists
  for row in insurance_data_reader:
    ages.append(float(row["age"]))
    sexes.append(row["sex"])
    BMIs.append(row["bmi"])
    children.append(row["children"])
    smokers.append(row["smoker"])
    regions.append(row["region"])
    charges.append(row["charges"])

  print(f"Ages: {ages} \nSexes: {sexes} \nBMIs: {BMIs} \nNumber of children: {children} \nSmoker status: {smokers} \nRegions: {regions} \nInsurance charges: {charges}")
```

```python
# Write function to calculate average insurance cost for a specific age

def average_insurance_by_age(analysis_age=0.0):
  print(f"analysis age is {analysis_age}")
  index = 0
  total_insurance_cost = 0.0
  count = 0
  for age in ages:
    # print(age)
    if age == analysis_age:
      print(f"age is equal!")
      count +=1
      total_insurance_cost = round(float(total_insurance_cost) + float(charges[index]), 2)
      average_insurance_cost = round(total_insurance_cost/count,2)
      print(f"index: {index} count: {count} new insurance cost: ${charges[index]} total_insurance_cost: ${total_insurance_cost} average_insurance_cost: ${average_insurance_cost}")
    else:
      if count == 0:
        average_insurance_cost = 0
    index+=1
  return(analysis_age, average_insurance_cost)
print(f"The average insurance costs for {average_insurance_by_age(37)[0]} year olds is ${average_insurance_by_age(37)[1]}")
```
